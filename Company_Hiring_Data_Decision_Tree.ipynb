```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Company_Hiring_DataSet.csv")
```

```python
df
```

```python
inputs = df.drop('Likely to Hire',axis='columns')
target = df['Likely to Hire']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LabelEncoder()
```

```python
le_Company_Size	= LabelEncoder()
le_Industry = LabelEncoder()
le_Revenue = LabelEncoder()
le_Location = LabelEncoder()
le_Employee_Turnover_Rate = LabelEncoder()
le_Employee_Skillset = LabelEncoder()
le_Job_Openings = LabelEncoder()
le_Recruitment_Budget = LabelEncoder()
le_Company_Age = LabelEncoder()
le_Growth_Rate = LabelEncoder()
```

```python
inputs['Company Size'] = le_Company_Size.fit_transform(inputs['Company Size'])
inputs['Industry'] = le_Industry.fit_transform(inputs['Industry'])
inputs['Revenue'] = le_Revenue.fit_transform(inputs['Revenue'])
inputs['Location'] = le_Location.fit_transform(inputs['Location'])
inputs['Employee Turnover Rate'] = le_Employee_Turnover_Rate.fit_transform(inputs['Employee Turnover Rate'])
inputs['Employee Skillset'] = le_Employee_Skillset.fit_transform(inputs['Employee Skillset'])
inputs['Job Openings'] = le_Job_Openings.fit_transform(inputs['Job Openings'])
inputs['Recruitment Budget'] = le_Recruitment_Budget.fit_transform(inputs['Recruitment Budget'])
inputs['Company Age'] = le_Company_Age.fit_transform(inputs['Company Age'])
inputs['Growth Rate'] = le_Growth_Rate.fit_transform(inputs['Growth Rate'])
inputs.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs,target)
```

```python
import numpy as np

# Your original 1D array
data = np.array([4,1,34,71,4,6,0,0,1,3])

# Reshape to 2D array for a single sample
data_2d = data.reshape(1, -1)

# Now you can use this 2D array for prediction
model.predict(data_2d)
```
